Creamos una función para la carga incremental de los datos extraídos de la API de Get On Board que tendrán siempre la misma estructura.

```python
def extraer_tabla(archivo):
    archivo_leido = pd.read_csv(archivo)
    archivo_leido = archivo_leido[["attributes.title","attributes.description","attributes.remote_modality","attributes.country","attributes.perks","attributes.min_salary","attributes.max_salary","attributes.company.data.attributes.name","attributes.company.data.attributes.country"]]
    archivo_leido = archivo_leido.reindex(columns=["attributes.title","attributes.description","attributes.company.data.attributes.name","attributes.company.data.attributes.country","attributes.country","attributes.remote_modality","attributes.perks","attributes.min_salary","attributes.max_salary"])
    archivo_leido = archivo_leido.rename({"attributes.company.data.attributes.country":"job_title",
                                                        "attributes.description": "requirements",
                                                        "attributes.company.data.attributes.name":"enterprise_name",
                                                        "attributes.company.data.attributes.country":"country_enterprise",
                                                        "attributes.country": "country_job",
                                                        "attributes.remote_modality":"remote_modality",
                                                        "attributes.perks":"perks",
                                                        "attributes.min_salary":"min_salary",
                                                        "attributes.max_salary":"max_salary"})
    archivo_final = archivo_leido.dropna()
    archivo_final = archivo_final.to_csv("archivo_final.csv")
    
    return archivo_final   
```

```python
extraer_tabla("ml.csv")
```
