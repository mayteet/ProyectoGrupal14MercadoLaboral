#### Bucle para sacar el nombre de la empresa en una nueva columna, que aplicamos a la columna 'reviews.val.reviewerInformation'.

```python
serie = []
for e in review_clean['reviews.val.reviewerInformation']:
    if type(e) == type(''):
        p = ''
        upper = e.split()
        cont = 0
        for i in upper:
            if i[0].isupper() and cont >= 4:
                p = p + ' ' + i
            else:
                cont = cont + 1
        serie.append(p)
    else:
        serie.append(np.nan)

review_clean['empresa'] = serie 
```
